```python
grades = [85, 92, 78, 90, 88, 95, 60, 72]
```

```python
#count grade
average_grade = sum(grades) /len(grades)
highest_grade = max(grades)
min_grade = min(grades)
```

```python
print(f"Total students: {len(grades)}")
print(f"Average Grade: {average_grade}")
print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {min_grade}")
```

```python
import matplotlib.pyplot as plt
#students name
students = ['mhoammed', 'fahad', 'omar', 'adam', 'ali', 'bader', 'malik', 'hakim' ]
#draw as line
plt.figure(figsize=(8, 5))
plt.bar(students, grades, color='skyblue')
#title
plt.title("Student Grades Analysis")
plt.xlabel("Students")
plt.ylabel("Grades")
plt.show()
```
